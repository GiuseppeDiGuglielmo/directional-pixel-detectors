```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("recon1.csv")
```

```python
df.head()
```

```python
X = df.values
X = np.reshape(X, (13,21))
```

```python
plt.figure(figsize = (16,5))
sns.heatmap(X, annot=False, fmt=".2g", cmap="YlGnBu")
plt.title('Recon. Cluster 1')
plt.savefig('reconCluster1.png')
```

```python
df2 = pd.read_csv("quantize1.csv")
```

```python
X1 = df2.values
X1 = np.reshape(X1, (13,21))
plt.figure(figsize = (16,5))
sns.heatmap(X1, annot=False, fmt=".2g", cmap="YlGnBu" )
plt.title('Quantized Cluster 1')
plt.savefig('quantizedCluster1.png')
```

```python
plt.figure(figsize = (16,5))
sns.heatmap(X, annot=True, fmt=".2g")
plt.xlim([5,16])
plt.ylim([2,10])
plt.title('Recon. Cluster 1')
plt.savefig('reconCluster1-annotated.png')
```

```python
X1 = df2.values
X1 = np.reshape(X1, (13,21))
plt.figure(figsize = (16,5))
sns.heatmap(X1, annot=False, fmt=".2g")
plt.xlim([5,16])
plt.ylim([2,10])
plt.title('Quantized Cluster 1')
plt.savefig('quantizedCluster1-annotated.png')
```
