```python
import numpy as np
import pandas as pd
from pandas import read_csv
import math
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('cut.csv')
```

```python
df1=df1.drop(['Unnamed: 0'],axis=1)
```

```python
df1=df1.drop(['z-entry'],axis=1)
df1
```

```python
df2=df1[df1['y-local'] < 1.5]
```

```python
df3 = df2[df2['y-local'] > 1.4]
df3
```

```python
df3 = df3.reset_index(drop=True)
```

```python
df4=df3[df3['pt'] > 2]
df4
```

```python
n = 62
df4=df4.iloc[:n]
df4
```

```python
df4 = df4.drop(['y-local'],axis=1)
df4 = df4.drop(['pt'], axis=1)
df4
```

```python
df4.to_csv('saved.csv', index=False)
```
