```python
import numpy as np
import pandas as pd
from pandas import read_csv
import math
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('highPTset128172.csv')
df2 = pd.read_csv('lowPTset128172.csv')
```

```python
#make high PT train set
df1.drop('z-entry', axis=1, inplace=True)
df1.drop('n_z', axis=1, inplace=True)
n = 102537
df1_Train_HighPT=df1.iloc[:n]
#df1_Train_HighPT
```

```python
df2.drop('z-entry', axis=1, inplace=True)
df2.drop('n_z', axis=1, inplace=True)
n = 102537
df1_Train_LowPT=df2.iloc[:n]
df1_Train_LowPT
```

```python
#get equal numbers of high PT and low PT samples, put together into one big training set
concatTrainSet = pd.concat([df1_Train_LowPT, df1_Train_HighPT])
```

```python
#shuffle data samples in train set
shuffled_df5 = concatTrainSet.sample(frac=1).reset_index(drop=True)
shuffled_df5
```

```python
shuffled_df5.to_csv('shuffled_d5.csv',index=False)
```

```python
def myfunc(x):
    if x < 0.3: 
        return 0
    else:
        return 1
    
shuffled_df5['pt'] = shuffled_df5['pt'].apply(myfunc)
shuffled_df5
```

```python
#these are the labels for the training set
shuffled_df5['pt'].to_csv('labels.csv',index=False)
```

```python
shuffled_df5.drop('y-local', axis=1, inplace=True)
shuffled_df5.drop('pt', axis=1, inplace=True)
shuffled_df5
```

```python
#apply 800 electron threshold
def counterNonZeroRows(X):
    countRows = 0
    for i in X:
        checkOne = False
        for j in i:
            if (j >= 800):
                checkOne = True
        if checkOne == True:
            countRows +=1
    return countRows
```

```python
list1 = []
for index, row in shuffled_df5.iterrows():
    X = row.values
    X = np.reshape(X,(13,21))
    nonZeroRows = counterNonZeroRows(X)
    list1.append(nonZeroRows)
```

```python
df_ySize = pd.DataFrame(list1)
df_ySize.columns = ['y-size']
df_ySize.to_csv('ysize1.csv',index=False)
```

```python
dfA = pd.read_csv('ysize1.csv')
dfA
```

```python
dfB = pd.read_csv('shuffled_d5.csv',usecols = ['y-local'])
dfB
```

```python
#now concatenate both dataframes since input will be: ysize and y_local
dfE = pd.concat([dfA,dfB],axis=1)
#dfE.to_csv('all.csv',index=False)
dfE
```

```python
dfE.to_csv('InputTrainingSet.csv',index=False)
```
