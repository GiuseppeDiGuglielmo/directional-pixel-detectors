```python
df1_train = pd.read_csv('InputTrainSet.csv')
df1_train.head()
#need to read in just the yprofile column, other things will stay the same
```

```python
df1_test = pd.read_csv('InputTestSet.csv')
df1_test.head()
```

```python
def replace(row):
  for i, item in enumerate(row):
     if row[i] >= -8 and row[i] <= -6.67:
        row[i] = 0
     elif ( row[i] >= -6.67 and row[i] <= -5.33):
        row[i] = 1 
     elif ( row[i] > -5.33 and row[i] <= -4):
        row[i] = 2
     elif ( row[i] > -4 and row[i] <= -2.67):
        row[i] = 3 
     elif ( row[i] > -2.67 and row[i] <= -1.33):
        row[i] = 4
     elif ( row[i] > -1.33 and row[i] <= 0):
        row[i] = 5 
     elif ( row[i] > 0 and row[i] <= 1.33):
        row[i] = 6
     elif ( row[i] > 1.33 and row[i] <= 2.67):
        row[i] = 7 
     elif ( row[i] > 2.67 and row[i] <= 4):
        row[i] = 8
     elif ( row[i] > 4 and row[i] <= 5.33):
        row[i] = 9 
     elif ( row[i] > 5.33 and row[i] <= 6.67):
        row[i] = 10
     else:
        row[i] = 11
  return row
```

```python
df1_train['y-local'].to_csv('y-localBeforeConvertingTrain.csv', index=None)
df1_test['y-local'].to_csv('y-localBeforeConvertingTest.csv', index=None)
```

```python
df2_train = pd.read_csv('y-localBeforeConvertingTrain.csv')
df2_train.head()
```

```python
df2_test = pd.read_csv('y-localBeforeConvertingTest.csv')
df2_test.head()
```

```python
df3_train = df2_train.progress_apply(lambda row : replace(row)) 
df3_train.head()
```

```python
df3_test = df2_test.progress_apply(lambda row : replace(row)) 
df3_test.head()
```

```python
df3_train.to_csv('y-localAfterConvertingTrain.csv', index=False)
df3_test.to_csv('y-localAfterConvertingTest.csv', index=False)
```
